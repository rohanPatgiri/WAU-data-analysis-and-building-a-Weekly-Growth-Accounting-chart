- Calculating Retained Users

```python
df = pd.read_csv("Active_Users2.csv")
```

```python
import pandas as pd

# Load your DataFrame (assuming it's already loaded into 'df')
# For demonstration, let's create a sample DataFrame


# Initialize a list to store the count of common entries for each week comparison
common_entries = []

# Iterate through each column (starting from the second column)
for i in range(1, df.shape[1]):
    # Get the current and previous columns
    current_col = df.columns[i]
    previous_col = df.columns[i - 1]
    
    # Drop NaN values and convert to sets for comparison
    current_set = set(df[current_col].dropna())
    previous_set = set(df[previous_col].dropna())
    
    # Find the common entries
    common_count = len(current_set.intersection(previous_set))
    
    # Append the result to the list
    common_entries.append(common_count)

# Create a DataFrame to store the results
retained_df = pd.DataFrame({
    'Week': df.columns[1:],  # Skip the first week as it has no previous week to compare
    'Retained': common_entries
})

# Display the results
#print(retained_df)

#standardizing the length of the dataframe
new_row = {'Week': 'w1', 'Retained': 0}
new_row_df = pd.DataFrame([new_row])
retained_df = pd.concat([new_row_df, retained_df], ignore_index=True)

print(retained_df)
```

Calculating Resurrected users

```python
df = pd.read_csv("Active_Users2.csv")
```

```python
import pandas as pd

# Sample DataFrame for demonstration


# Initialize a list to store the count of resurrected users for each week
resurrected_users = []

# Iterate through each week's column (starting from the third week)
for i in range(2, df.shape[1]):
    # Get the current and previous columns
    current_col = df.columns[i]
    previous_col = df.columns[i - 1]
    
    # Drop NaN values and convert to sets for comparison
    current_set = set(df[current_col].dropna())
    previous_set = set(df[previous_col].dropna())
    
    # Initialize an empty set for all preceding weeks' users
    preceding_users = set()
    
    # Accumulate all users from week 1 to week (n-2)
    for j in range(0, i - 1):
        preceding_users.update(set(df[df.columns[j]].dropna()))
    
    # Resurrected users are those who are in the current week but not in the previous week,
    # but were active in any of the preceding weeks.
    resurrected_set = current_set - previous_set
    resurrected_set = resurrected_set & preceding_users
    
    # Find the count of resurrected users
    resurrected_count = len(resurrected_set)
    
    # Append the result to the list
    resurrected_users.append(resurrected_count)

# Create a DataFrame to store the results
resurrected_df = pd.DataFrame({
    'Week': df.columns[2:],  # Skip the first two weeks as resurrection calculation is not possible
    'Resurrected': resurrected_users
})

# Display the results
#print(resurrected_df)

#standardizing the length of the dataframe
new_row = {'Week': 'w1', 'Resurrected': 0}
new_row_df = pd.DataFrame([new_row])
ressurrected_df = pd.concat([new_row_df, resurrected_df], ignore_index=True)


```

```python
 #Concatenate DataFrames
retained_df = retained_df[['Retained']]
churn_df = churn_df[['Churn']]

# Use `axis=1` to concatenate along columns
#oncatenated_df = pd.concat([wau_df, retained_df, churn_df], axis=1)

# Display the concatenated DataFrame
#print(concatenated_df)

#.................
# Add missing rows with zeros to resurrected_df
# Create a full-length DataFrame filled with zeros
full_length = len(wau_df)
zeros_df = pd.DataFrame({'Resurrected': [0] * full_length})

# Update the zeros_df with actual values from resurrected_df starting from the third row
zeros_df.iloc[2:] = resurrected_df[['Resurrected']].reset_index(drop=True)

# Concatenate DataFrames
# Use `axis=1` to concatenate along columns
concatenated_df = pd.concat([wau_df, retained_df, churn_df, zeros_df], axis=1)

# Display the concatenated DataFrame
print(concatenated_df)

```

```python

# Calculate the 'New' column
concatenated_df['New'] = concatenated_df['WAU'] - concatenated_df['Retained'] - concatenated_df['Resurrected']

# Display the concatenated DataFrame with the new column
print(concatenated_df)
```

- Calculating Retention Rate

```python

# Calculate the 'Retention Rate' as Retained / Previous_WAU
concatenated_df['Retention Rate'] = concatenated_df['Retained'] / concatenated_df['WAU'].shift(1)

# Optionally, drop the 'Previous_WAU' column if it's no longer needed
# concatenated_df.drop(columns=['Previous_WAU'], inplace=True)

# Display the updated DataFrame
print(concatenated_df)
```

- Calculating Quick Ratio

```python
concatenated_df["Quick Ratio"] = (concatenated_df["New"]+ concatenated_df["Resurrected"])/ concatenated_df["Churn"]
print(concatenated_df)
```

```python
# Replacing "NaN" and "inf" with 0
import numpy as np
# Replace 'inf' values in 'Quick Ratio' column with 0
concatenated_df['Quick Ratio'] = concatenated_df['Quick Ratio'].replace(np.inf, 0)

# Replace 'NaN' values in 'Retention Rate' column with 0
concatenated_df['Retention Rate'] = concatenated_df['Retention Rate'].fillna(0)

wau_growth_data = concatenated_df
# Display the updated DataFrame
print(wau_growth_data)
```

```python
import pandas as pd
from fpdf import FPDF

# Load the CSV file into a DataFrame
df = pd.read_csv('wauGrowthData.csv')

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'WAU Growth Data', 0, 1, 'C')

    def table(self, data):
        self.set_font('Arial', '', 10)
        col_width = self.epw / len(data.columns)  # Equal width for all columns
        row_height = self.font_size * 1.5

        # Add header
        for col in data.columns:
            self.cell(col_width, row_height, col, border=1)
        self.ln(row_height)

        # Add rows
        for index, row in data.iterrows():
            for item in row:
                self.cell(col_width, row_height, str(item), border=1)
            self.ln(row_height)

# Create a PDF document
pdf = PDF()
pdf.add_page()
pdf.table(df)

# Save the PDF
pdf_file = 'wau_growth_data.pdf'
pdf.output(pdf_file)
print(f'PDF document has been successfully saved as {pdf_file}')
```
